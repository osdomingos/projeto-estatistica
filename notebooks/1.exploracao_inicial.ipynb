<strong>Artistas:</strong>

```python
df_artistas.columns
```

```python
dict_variaveis_df_artistas = {
    "Variáveis": [],
    "Tipo de dado": [],
    "Descrição": [],
    "Tipo de variável": [],
    "Ações": []
}

for column in df_artistas.columns:
    dict_variaveis_df_artistas["Variáveis"].append(column)
    dict_variaveis_df_artistas["Tipo de dado"].append(type(column).__name__)
    dict_variaveis_df_artistas["Descrição"].append('')
    dict_variaveis_df_artistas["Tipo de variável"].append('')
    dict_variaveis_df_artistas["Ações"].append('')
```

```python
dict_variaveis_df_artistas
```

```python
df_artistas_variaveis = pd.DataFrame(data=dict_variaveis_df_artistas)
```

```python
descricoes = [
    "'Chave estrangeira' que relaciona Artistas com Obras",
    "Nome do artista/organização",
    "Nacionalidade e data de nasc. - falecimento",
    "Nacionalidade",
    "Gênero",
    "Ano de nascimento",
    "Ano de falecimento",
    "ID de busca na WikiData",
    "ID da Union List of Artist Names® da Getty Research Institue"
]

tipos_de_variaveis = [
    "numérica discreta",
    "categórica nominal",
    "categórica nominal",
    "categórica nominal",
    "categórica nominal",
    "numérica discreta",
    "numérica discreta",
    "categórica nominal",
    "categórica nominal"
]

acoes = [
    "Converter em int",
    "Nenhuma",
    "Remover: outras colunas dão conta da mesma informação de forma isolada e de mais fácil leitura e manipulação",
    "Alterar NaN para 'Nationality unknown', que já está presente no dataset",
    "Investigar se há outra indicação de gênero além de 'male' ou 'female'",
    "Converter em int para poder fazer operações de idade",
    "Converter em int para poder fazer operações",
    "Converter em URL para acesso rápido 'https://www.wikidata.org/wiki/<Wiki QID>'",
    "Primeiro remover o '.0', que não retorna nada, então converter em URL para acesso rápido 'https://www.getty.edu/vow/ULANFullDisplay?find=<ULAN>&role=&nation=&prev_page=1&subjectid=<ULAN>'"
]

for i in df_artistas_variaveis.index:
    df_artistas_variaveis.at[i, 'Descrição'] = descricoes[i]
    df_artistas_variaveis.at[i, 'Tipo de variável'] = tipos_de_variaveis[i]
    df_artistas_variaveis.at[i, 'Ações'] = acoes[i]
```
